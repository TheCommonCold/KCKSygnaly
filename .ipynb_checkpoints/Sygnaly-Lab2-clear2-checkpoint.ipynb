```python
from __future__ import division
from pylab import *
from numpy import *
from scipy import *
from random import *

from ipywidgets import *
import math as mt
import cmath

# funkcja usuwająca częstotliwości
def removeFrequencies(fftNum, freq, amp, rem):
    for r in rem:
        for i, f in enumerate(freq):
            if f == r:
                fftNum[i] = complex(0.0, 0.0)
                fftNum[-i] = complex(0.0, 0.0)
                amp[i] = 0
                amp[-i] = 0
    return fftNum, amp
                

# Funkcja licząca p
def getSignalParameters(samples, samplingFreq):
    numberOfSamples = len(samples)
    
    complexNumbers = fft(samples)
    phaseShifts = [degrees(cmath.phase(cnumber)) for cnumber in complexNumbers]
    amplitudes = abs(complexNumbers) / (numberOfSamples/2)
    amplitudes[0] /= 2
    
    harmonicFreqs = [(i/numberOfSamples)*samplingFreq for i in range(int(numberOfSamples))]
    
    return complexNumbers, harmonicFreqs, amplitudes, phaseShifts

    
```

array = genfromtxt(‘spots.txt’)  (lub array=[float(x) for x in open('spots.txt').read().split()])

# Zadanie 2 - proste filtrowanie

Wykreśl sygnał $sin(2*pi*t) + sin(4*pi*t)$, T=1s, w=20Hz. Za pomocą FFT, przekształć sygnał do dziedziny częstotliwości. Następnie usuń składowe o częstotliwości 2Hz. Tak zmodyfikowany sygnał przekształć do dziedziny czasu i wykreśl go.

```python
samplingFreq = 20
samplingPeriod = 1

fig = plt.figure(figsize=(15, 15), dpi=80)


cont_t = np.arange(0, samplingPeriod, 1.0/200.0)
discrete_t = np.arange(0, samplingPeriod, 1.0/samplingFreq)
func1 = lambda t : sin(2*pi*t)
func2 = lambda t : sin(4*pi*t)
samples = sum([func1(discrete_t), func2(discrete_t)] , axis=0)
resultant = sum([func1(cont_t), func2(cont_t)] , axis=0)

ax = fig.add_subplot(221)
ax.plot(cont_t, func1(cont_t), linestyle='--')
ax.plot(cont_t, func2(cont_t), linestyle='--')
ax.plot(cont_t, resultant, linestyle='-', color='red')
ax.plot(discrete_t, samples, 'o')

fftNumbers, freqs, amps, shifts = getSignalParameters(samples, samplingFreq)

ax = fig.add_subplot(222)
ax.stem(freqs, amps, '-')

modFFT, amps = removeFrequencies(fftNumbers, freqs, amps, [2])
modifiedsamples = ifft(modFFT).real

ax = fig.add_subplot(223)
ax.plot(discrete_t, modifiedsamples, "o-")

ax = fig.add_subplot(224)
ax.stem(freqs, amps, '-')
```

# Zadanie 3 - Informacja o fazie

Wykreśl sygnał $sin(2*pi*t) + sin(4*pi*t)$, T=1s, w=20Hz. Tym razem oprócz spektrum, wykreśl wykres z informacją o fazie poszczególnych częstotliwości (faza = arg(z)), gdzie z=a+bi. To samo wykonaj dla sin(2*pi*t) + cos(4*pi*t). Porównaj otrzymane wykresy. Przydatna może być funkcja angle (z parametrem deg=True)

```python
samplingFreq = 20
samplingPeriod = 1

fig = plt.figure(figsize=(15, 15), dpi=80)


cont_t = np.arange(0, samplingPeriod, 1.0/200.0)
discrete_t = np.arange(0, samplingPeriod, 1.0/samplingFreq)
func1 = lambda t : sin(2*pi*t)
func2 = lambda t : sin(4*pi*t)
samples = sum([func1(discrete_t), func2(discrete_t)] , axis=0)
resultant = sum([func1(cont_t), func2(cont_t)] , axis=0)

ax = fig.add_subplot(421)
ax.plot(cont_t, func1(cont_t), linestyle='--')
ax.plot(cont_t, func2(cont_t), linestyle='--')
ax.plot(cont_t, resultant, linestyle='-', color='red')
ax.plot(discrete_t, samples, 'o')

fftNumbers, freqs, amps, shifts = getSignalParameters(samples, samplingFreq)

ax = fig.add_subplot(422)
ax.stem(freqs, amps, '-')


fftNumbers = [(num if abs(imag(num)) > 0.01 else 0) for num in fftNumbers]

ax = fig.add_subplot(423)
ax.stem(freqs, angle(fftNumbers, True))



func1 = lambda t : sin(2*pi*t)
func2 = lambda t : cos(4*pi*t)
samples = sum([func1(discrete_t), func2(discrete_t)] , axis=0)
resultant = sum([func1(cont_t), func2(cont_t)] , axis=0)

fftNumbers, freqs, amps, shifts = getSignalParameters(samples, samplingFreq)

ax = fig.add_subplot(425)
ax.plot(cont_t, func1(cont_t), linestyle='--')
ax.plot(cont_t, func2(cont_t), linestyle='--')
ax.plot(cont_t, resultant, linestyle='-', color='red')
ax.plot(discrete_t, samples, 'o')

ax = fig.add_subplot(426)
ax.stem(freqs, amps, '-')

fftNumbers = [(num if abs(imag(num)) > 0.01 else 0) for num in fftNumbers]

ax = fig.add_subplot(427)
ax.stem(freqs, angle(fftNumbers, True))
```

Wczytaj plik err.wav. Wykreśl jego spektrum. Spróbuj także skali logarytmicznej. Określ dominujące w sygnale częstotliwości. Przydatne:

import scipy.io.wavfile

w, signal = scipy.io.wavfile.read('err.wav')

signal = [s[0] for s in signal]   #Tylko pierwszy kanał

yscale(‘log’)

spectrum[::10]        # co 10-ty element


```python
import scipy.io.wavfile

fig = plt.figure(figsize=(15, 15), dpi=80)
samplingFreq = 44100

w, signal = scipy.io.wavfile.read('err.wav')
signal = [s[0] for s in signal] #Tylko pierwszy kanał
fftNumbers, freqs, amps, shifts = getSignalParameters(signal[::10], samplingFreq)


ax = fig.add_subplot(111)
plt.yscale('log')
ax.stem(freqs, amps, '-')
```
